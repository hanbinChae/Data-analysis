```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4],[4,3,2,1],'o-r') 
#y축 x축 순서 ex(1,4),(2,3)...
#'marker,linestyle,color'
plt.ylabel('y-label')
plt.xlabel('x-label')
# plt.axis() 축의 범위 지정 [xmin,xmax,ymin,ymax]
plt.show()
```

```python
import numpy as np
t = np.arange(0.,5.,0.2) #200ms 간격으로 균일하게 샘플된 시간
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16],'yp-.')
plt.xlabel('X-Label',)
plt.ylabel('Y-Label')
plt.axis([0, 5, 0, 20]) 
plt.show()
```

```python
plt.plot([1,4,9,16],'r')
plt.plot([1,2,9,16],'b')
plt.plot([1,4,6,16],'g')
# 인자 종류
# color, marker, linestyle
plt.show()
```

```python
x= [1,2,3,4] 
y = [1,4,9,16]

plt.plot(x,y)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_between(x[1:3],y[1:3],alpha=0.5)
plt.fill_betweenx(y[2:4], x[2:4], color='pink', alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_between(x[1:3], y1[1:3], y2[1:3], color='lightgray', alpha=0.5)

plt.show()
```
