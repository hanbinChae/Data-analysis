문제 22

```python
low_prices = [100, 200, 400, 800, 1000]
high_prices = [150, 300, 430, 880, 1000]
volatility = []
a=0
for i in high_prices:
  volatility.append((i-low_prices[a]))
  a+=1
print(volatility)
```
